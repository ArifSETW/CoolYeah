```python
def bellman_ford(graph, source):
  """
  Menghitung jarak terpendek dari source ke semua node lain dalam graph.

  Args:
    graph: Graf berbobot.
    source: Node sumber.

  Returns:
    Daftar jarak terpendek dari source ke semua node lain.
  """

  # Inisialisasi jarak terpendek
  distances = [float("inf") for _ in range(len(graph))]
  distances[source] = 0

  # Iterasi sebanyak jumlah node
  for _ in range(len(graph)):
    # Periksa semua sisi
    for u, v, weight in graph.edges:
      # Perbarui jarak terpendek jika ada jalur yang lebih pendek
      if distances[u] + weight < distances[v]:
        distances[v] = distances[u] + weight

  return distances
```

```python
import matplotlib.pyplot as plt
import networkx as nx

graph = {
    "A": {"C": 1, "B": 2, "E": 3, "D": 4},
    "B": {"C": 3,},
    "C": {"D": 2},
    "D": {"C": 3},
    "E": {"D": 2}
}

# Menentukan posisi node secara manual
pos = {
    "A": (0, 0),
    "B": (-0.951, -0.309),
    "C": (-0.588, -0.809),
    "D": (0.588, -0.809),
    "E": (0.951, -0.309)
}

# Menampilkan gambar graf
G = nx.Graph(graph)
nx.draw(G, pos, with_labels=True, node_size=700, node_color='green', font_size=13, font_color='white', connectionstyle='arc3,rad=0.1')

# Menambahkan label bobot pada tepi
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Bellman-Ford")
plt.show()
```
